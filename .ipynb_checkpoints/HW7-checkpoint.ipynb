```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
from citipy import citipy
```

```python
#API key
api_key="61e7b020d7bd7f993be4531548ed6296"
```

```python
#Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
# Output File (CSV)
output_data_file = "output_data/cities.csv"
#list for holding lat_lngs and cities
lat_lngs = []
cities = []
#create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000,size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size =1500)
lat_lngs=zip(lats, lngs)
#identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city=citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    #if the city is unique, then add it to our cities list
    if city not in cities:
        cities.append(city)
#print the city count to confirm sufficient count
len(cities)
```

```python
# create variables for collecting data from api
city =[]
cloudness=[]
country=[]
date = []
humidity = []
lat = []
lng = []
temp = []
wind_speed=[]
#set up counter as 1
counter = 1
# create query_url based on base_url and an example from weather api documentation.
base_url = "http://api.openweathermap.org/data/2.5/weather?"
api_key="61e7b020d7bd7f993be4531548ed6296"
print("Beginning Data Retrieval")
print("------------------------")
for cityname in cities:
    query_url = base_url + "appid=" + api_key + "&q=" + cityname
    print(query_url)
    response = requests.get(query_url).json()
    #using pretty print to get location for each variable in json file, 
    #so I can collect data from response.
    #print(json.dumps(response, indent =4, sort_keys = True))
    counter += 1
    try:
        city.append(response["name"])
        cloudness.append(response["clouds"]["all"])
        country.append(response["sys"]["country"])
        date.append(response["dt"])
        humidity.append(response["main"]["humidity"])
        lat.append(response["coord"]["lat"])
        lng.append(response["coord"]["lon"])
        temp.append(response["main"]["temp"])
        wind_speed.append(response["wind"]["speed"])
        print("Proceesing City # " ,counter, ' | ' , cityname)
    except: 
        print("City not found. Skipping...")
    
    if counter > 50:
       time.sleep(60)
       counter = 1
```

```python
cities_d = {"City": city,
            "Cloudness": cloudness,
            "Country": country,
            "Date": date,
            "Humidity":humidity,
            "Lat":lat,
            "Lng": lng, 
            "Temp": temp,
            "Wind Speed": wind_speed
           }
cities_weather = pd.DataFrame(cities_d)
cities_weather.head()
```
